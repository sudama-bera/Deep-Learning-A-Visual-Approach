```python
import numpy as np
import matplotlib.pyplot as plt
import math
import seaborn as sns ; sns.set()
```

```python
# Simple, self‑contained setup for file handling – no File_Helper needed

import os

save_files = False  # set True if you actually want to write files

# Base directories (match what you've used elsewhere)
input_data_directory = os.path.join(os.getcwd(), "input_data")
output_data_directory = os.path.join(os.getcwd(), "saved_output")

os.makedirs(input_data_directory, exist_ok=True)
os.makedirs(output_data_directory, exist_ok=True)

print("Input data directory :", input_data_directory)
print("Output data directory:", output_data_directory)
print("save_files =", save_files)
```

```python
def make_noise_figure(filename=None):
    noise = np.random.uniform(low=-1, high=1, size=(224, 224, 3))
    plt.figure(figsize=(8, 3))

    plt.subplot(1, 4, 1)
    plt.imshow(noise)
    plt.xticks([], [])
    plt.yticks([], [])
    plt.title("Noise image", fontsize=16)

    plt.subplot(1, 4, 2)
    plt.imshow(noise[:, :, 0])
    plt.xticks([], [])
    plt.yticks([], [])
    plt.title("Red channel", fontsize=16)

    plt.subplot(1, 4, 3)
    plt.imshow(noise[:, :, 1])
    plt.xticks([], [])
    plt.yticks([], [])
    plt.title("Green channel", fontsize=16)

    plt.subplot(1, 4, 4)
    plt.imshow(noise[:, :, 2])
    plt.xticks([], [])
    plt.yticks([], [])
    plt.title("Blue channel", fontsize=16)

    plt.tight_layout()

    # optional saving via plain matplotlib, since file_helper is gone
    if filename and save_files:
        out_path = os.path.join(output_data_directory, filename)
        plt.savefig(out_path)
        print("Saved noise figure to:", out_path)

    plt.show()
```

```python
make_noise_figure('noise-input')
```

```python
import math
import numpy as np

def make_blob_in_box(filter_size, blob_x, blob_y, blob_r):
    grid = np.zeros((filter_size, filter_size))
    for y in range(filter_size):
        for x in range(filter_size):
            dx = x - blob_x
            dy = y - blob_y
            r = math.sqrt(dx * dx + dy * dy)
            v = math.exp(-r * r / blob_r)
            grid[y, x] = v
    return grid
```

```python
def draw_grid(grid_size):
    for i in range(grid_size):
        plt.plot([i + 0.5, i + 0.5], [-0.5, grid_size - 0.5],
                 color="black", lw=0.4)
        plt.plot([-0.5, grid_size - 0.5], [i + 0.5, i + 0.5],
                 color="black", lw=0.4)
```

```python
def draw_filter_set(filter_size, gram_size, f1, f2, f1_bumps, f2_bumps, filename=None):
    def draw_grid_with_title(grid, title):
        plt.imshow(grid, cmap="cool", interpolation="nearest")
        plt.title(title, fontsize=13)
        plt.xticks([], [])
        plt.yticks([], [])
        draw_grid(grid_size)

    grid_size = 8

    # Build first filter from bumps
    filter1 = np.zeros((filter_size, filter_size))
    for bump in f1_bumps:
        grid = make_blob_in_box(grid_size, bump["x"], bump["y"], bump["r"])
        filter1 += grid

    # Build second filter from bumps
    filter2 = np.zeros((filter_size, filter_size))
    for bump in f2_bumps:
        grid = make_blob_in_box(grid_size, bump["x"], bump["y"], bump["r"])
        filter2 += grid

    product = filter1 * filter2
    product_sum = np.sum(product)

    plt.figure(figsize=(10, 3))

    # Filter 1
    plt.subplot(1, 4, 1)
    plt.imshow(filter1, cmap="cool", interpolation="nearest")
    title = f"Filter{f1:d} map ({np.min(filter1):.2f}-{np.max(filter1):.2f})"
    plt.title(title, fontsize=13)
    plt.xticks([], [])
    plt.yticks([], [])
    draw_grid(grid_size)

    # Filter 2
    plt.subplot(1, 4, 2)
    title = f"Filter{f2:d} map ({np.min(filter2):.2f}-{np.max(filter2):.2f})"
    draw_grid_with_title(filter2, title)

    # Product
    plt.subplot(1, 4, 3)
    title = f"Product ({np.min(product):.2f}-{np.max(product):.2f})"
    draw_grid_with_title(product, title)

    # Gram matrix entry
    plt.subplot(1, 4, 4)
    gram = np.zeros((gram_size, gram_size))
    gram[f1, f2] = gram[f2, f1] = product_sum
    plt.imshow(gram, cmap="Reds", interpolation="nearest", vmin=0, vmax=10.9)
    title = f"Gram[{f1:d},{f2:d}]={product_sum:.2f}"
    plt.title(title, fontsize=13)
    plt.xticks([], [])
    plt.yticks([], [])
    draw_grid(gram_size)

    # Optional save without file_helper
    if filename and save_files:
        out_path = os.path.join(output_data_directory, filename)
        plt.savefig(out_path)
        print("Saved filter set figure to:", out_path)

    plt.show()
```

```python
# filter responses
f1 = [{"x": 3, "y": 4, "r": 6}]
f2 = [{"x": 3, "y": 5, "r": 10}]
f3 = [{"x": 7, "y": 2, "r": 5}]
f4 = [
    {"x": 6, "y": 6, "r": 3},
    {"x": 1, "y": 4, "r": 2.5},
    {"x": 5, "y": 1, "r": 2},
]
f5 = [{"x": 4, "y": 5, "r": 10}]
```

```python
draw_filter_set(filter_size=8, gram_size=6, f1=1, f2=2,
                f1_bumps=f1, f2_bumps=f2,
                filename='gram-filters-12')
```
